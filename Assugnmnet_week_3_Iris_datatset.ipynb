Visualization of Iris Dataset with the help of panda and numoy libraries

```python
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
```

# 1. LOAD DATASET

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, names=names)
```

# 2. Summarize the Dataset

```python
#shape
print(dataset.shape)
```

# 3. Peek The Data

```python
print(dataset.tail())
```

# 4.Head The Data

```python
print(dataset.head())
```

# Statistical Summary

```python
print(dataset.describe())
```

# Class Distribution
Letâ€™s now take a look at the number of instances (rows) that belong to each class. We can view this as an absolute count.


```python
print(dataset.groupby('class').size())
```

# 4. Data Visualization
We now have a basic idea about the data. We need to extend that with some visualizations.

We are going to look at two types of plots:

<li>Univariate plots to better understand each attribute.
<li>Multivariate plots to better understand the relationships between attributes.


# 4.1 Univariate Plots

```python
dataset.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

```python
dataset.hist()
plt.show()
```

# 4.2 Multivariate Plots

```python
# scatter plot matrix
scatter_matrix(dataset)
plt.show()
```

```python
sepal_length=dataset['sepal-length']
sepal_length.plot(kind='pie',figsize=(7,8))
```
